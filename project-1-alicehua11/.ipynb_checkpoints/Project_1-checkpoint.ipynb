# Project 1: Part 3

### Question 1: What are 5 most popular trips that you would call "commuter trips"?

Definition: a commuter trip is a trip that subscribers (members) regularly take in Morning to Mid Morning (6am - 10am) and return in the Afternoon (5pm - 7pm) hours and that they would return to the same station they started with as one complete trip. However, in this dataset, we don't have specific data about each rider to ensure that they've completed one round trip from 2 different rides, so we will assume that if the start station and end station are different, then it will be counted as a commuter trip along with other criteria mentioed above.<br>
Note: Different from **Part 1** in querying for the 10 most popular locations for commuter trips, this answer include only subscribers as riders of commuter trips, and total number of trips per station pair is more than 100. These stronger assumptions are made because I noticed later on in the queries that subscribers tend to ride on average 10 min, and these short trips happened within the intervals of commuting time defined above. <br>
Below is a table of 5 distinct stations for 5 most popular trip.

````python
# Manipulating dataframe to get the right format
commuter_trips = pd.read_csv('commuter_trips1.csv')
commuter_trips_agg = commuter_trips.groupby(
    ['start_station_name', 'city']).agg(
    {'avg_duration': ['mean'], 'number_of_trips':['sum']}).reset_index()
commuter_trips_agg.columns = ["_".join(x) for x in commuter_trips_agg.columns.ravel()]
commuter_trips_agg = commuter_trips_agg.sort_values('number_of_trips_sum',ignore_index=True, ascending=False).head(10)
commuter_trips_agg.head()
````

A visualization for 5 most popular trips compared to 5 other trips.

````python
f, ax = plt.subplots(figsize = (15,8), dpi=300)
sns.barplot(data = commuter_trips_agg, y='start_station_name_', x='number_of_trips_sum', label='start_station_name_',palette = 'colorblind')
# sns.set_context('paper')
ax.set_ylabel('')    
ax.set_xlabel('Number of Trips', fontsize=15)
ax.set_title("Most Popular Commuting Trips (Top 5 included)",  fontsize=15)
````

The top 10 trips are all from San Francisco including our top 5 most popular commuter trips. This tells us San Francisco is a hot market for Lyft bikes. The first 2 stations of these trips confirm that these are commuter trips since many of trips are made starting at transportation hubs like Caltrain and Transbay Terminal. <br>
But, what about the other cities? Let's take a look at the total number of commuter trips for each city for each day of the week. We have 5 cities: San Francisco, San Jose, Mountain View, Palo Alto, Redwood City and 5 days of the work week: Monday-Friday.

````python
#Plotting for subscriber commuting trips
ax = sns.catplot(x='dow_str', y='number_of_trips',
                 col='city', aspect=.9,
                 col_wrap=3, kind='bar', ci=None,
                 hue='start_hour_str', height=10, 
                 palette='deep',
                 data=commuter_trips)
(ax.set_xticklabels(['Monday','Tuesday','Wednesday', 'Thursday','Friday'])
    .set(ylim=(0, 1000))
    .set(ylabel="Number of Trips"))
ax.fig.autofmt_xdate()
for ax in ax.axes.flatten():
    ax.tick_params(labelbottom=True)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.subplots_adjust(hspace=0.7, wspace=0.4)
````

The graphs above show us that *Lyft* bikes are not popular in cities like Palo Alto and Redwood city. The number of riders for San Francisco is consistently high across the weekdays, partiticularly in the Morning and Afternoon coinciding with work start/end hours and tapering toward Friday when some people might not have work perhaps because it's closer to the weekend. San Jose show a very consistent number of trips across the weekdays as well, Mondays seem to be the busiest day for most cities meaning the subscribers are often riding to their workplaces as the workweek starts.

### APPENDIX for BigQuery Views

### Commuter Trip Views

```sql
SELECT *
FROM `bigquery-public-data.san_francisco.bikeshare_trips` regulartable
INNER JOIN (SELECT trip_id AS id2, start_date AS sdate,
       EXTRACT(DAYOFWEEK FROM start_date) AS dow_int,
       CASE EXTRACT(DAYOFWEEK FROM start_date)
           WHEN 1 THEN "Sunday"
           WHEN 2 THEN "Monday"
           WHEN 3 THEN "Tuesday"
           WHEN 4 THEN "Wednesday"
           WHEN 5 THEN "Thursday"
           WHEN 6 THEN "Friday"
           WHEN 7 THEN "Saturday"
           END AS dow_str,
       CASE 
           WHEN EXTRACT(DAYOFWEEK FROM start_date) IN (1, 7) THEN "Weekend"
           ELSE "Weekday"
           END AS dow_weekday,
       EXTRACT(HOUR FROM start_date) AS start_hour,
       CASE 
           WHEN EXTRACT(HOUR FROM start_date) <= 5  OR EXTRACT(HOUR FROM start_date) >= 23 THEN "Nightime"
           WHEN EXTRACT(HOUR FROM start_date) >= 6 and EXTRACT(HOUR FROM start_date) <= 8 THEN "Morning"
           WHEN EXTRACT(HOUR FROM start_date) >= 9 and EXTRACT(HOUR FROM start_date) <= 10 THEN "Mid Morning"
           WHEN EXTRACT(HOUR FROM start_date) >= 11 and EXTRACT(HOUR FROM start_date) <= 13 THEN "Mid Day"
           WHEN EXTRACT(HOUR FROM start_date) >= 14 and EXTRACT(HOUR FROM start_date) <= 16 THEN "Early Afternoon"
           WHEN EXTRACT(HOUR FROM start_date) >= 17 and EXTRACT(HOUR FROM start_date) <= 19 THEN "Afternoon"
           WHEN EXTRACT(HOUR FROM start_date) >= 20 and EXTRACT(HOUR FROM start_date) <= 22 THEN "Evening"
           END AS start_hour_str
FROM `bigquery-public-data.san_francisco.bikeshare_trips`) dayofweek 
ON regulartable.trip_id = dayofweek.id2
```

### Bike City Year (total bikes per city per year per station) View

```sql
SELECT station_id, city, year, total_bikes
FROM(
  SELECT DISTINCT tb1.station_id, year, total_bikes, city
  FROM (
    SELECT station_id, total_bikes, 
    EXTRACT (YEAR from time) AS year
    FROM `teak-radio-287600.bike_trip_data.total_bike_station_ordered`
    ) tb1
   INNER JOIN 
   `teak-radio-287600.bike_trip_data.city_with_stationid_trips` tb2
   ON tb1.station_id = tb2.station_id
   )
ORDER BY year
```

### Trips with Datetime Transformed View

```sql
SELECT *
FROM (
  SELECT *
  FROM `bigquery-public-data.san_francisco.bikeshare_trips` regulartable
  INNER JOIN (
            SELECT id2, start_date as sdate, dow_str, dow_weekday, start_hour_str
            FROM `teak-radio-287600.bike_trip_data.dayofweek`) dayofweek 
  ON regulartable.trip_id = dayofweek.id2) tablewithdates
INNER JOIN (
      SELECT trip_id AS id3, duration_sec AS sec, 
       CAST(ROUND(duration_sec / 60.0) AS INT64) AS duration_minutes,
       CAST(ROUND(duration_sec / 3600.0) AS INT64) AS duration_hours_rounded,
       ROUND(duration_sec / 3600.0, 1) AS duration_hours_tenths
      FROM `bigquery-public-data.san_francisco.bikeshare_trips`) tablewithtime
ON tablewithdates.id2 = tablewithtime.id3
```
